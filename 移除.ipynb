```python
import os

def delete_pdb_files(folder_path):
    # 检查路径是否存在
    if not os.path.exists(folder_path):
        print("指定的文件夹不存在")
        return

    # 遍历文件夹
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.pdb'):
                # 构建完整的文件路径
                file_path = os.path.join(root, file)
                # 删除文件
                os.remove(file_path)
                print(f"已删除文件：{file_path}")

# 使用示例
folder_path = 'biqinPDB'  # 替换为你的文件夹路径
delete_pdb_files(folder_path)
```

```python
import os

# 设置目标文件夹路径
folder_path = 'biqinPDB'

# 遍历文件夹中的所有文件
for filename in os.listdir(folder_path):
    # 检查文件名是否以 'processed_' 开头
    if filename.startswith('processed_'):
        # 生成新的文件名
        new_filename = filename[len('processed_'):]
        
        # 构造完整的文件路径
        old_file_path = os.path.join(folder_path, filename)
        new_file_path = os.path.join(folder_path, new_filename)
        
        # 重命名文件
        os.rename(old_file_path, new_file_path)
        print(f'Renamed: {filename} -> {new_filename}')
```

# 将文件夹中的sdf文件都转换成mol2

```python
import os
import subprocess

def convert_sdf_to_mol2(sdf_file, mol2_file):
    """
    使用Open Babel将SDF文件转换为MOL2格式。

    参数：
    sdf_file (str): 输入SDF文件路径。
    mol2_file (str): 输出MOL2文件路径。
    """
    command = f"obabel {sdf_file} -O {mol2_file}"
    subprocess.run(command, shell=True, check=True)
    print(f"已将 {sdf_file} 转换为 {mol2_file}")

def convert_all_sdf_in_folder(input_folder, output_folder):
    """
    批量将指定文件夹中的所有SDF文件转换为MOL2格式，并保存到另一个文件夹。

    参数：
    input_folder (str): 包含SDF文件的输入文件夹路径。
    output_folder (str): 用于保存转换后的MOL2文件的输出文件夹路径。
    """
    # 确保输出文件夹存在
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 遍历输入文件夹中的所有文件
    for filename in os.listdir(input_folder):
        if filename.endswith(".sdf"):
            sdf_file = os.path.join(input_folder, filename)
            mol2_file = os.path.join(output_folder, filename.replace(".sdf", ".mol2"))
            convert_sdf_to_mol2(sdf_file, mol2_file)

# 示例用法
input_folder = "./input_ligand"  # 输入文件夹路径
output_folder = "output_ligand"  # 输出文件夹路径

convert_all_sdf_in_folder(input_folder, output_folder)
```
